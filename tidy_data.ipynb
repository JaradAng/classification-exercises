Attendance Data

Read the data from the attendance table and calculate an attendance percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

You should end up with something like this:

```python
df_attendance = pd.read_csv('attendance.csv')

df_attendance.head()
```

```python
df_attendance = df_attendance.rename(columns={'Unnamed: 0': 'Name'})
```

```python
attendance_melt = df_attendance.melt(id_vars=['Name'], var_name='date', value_name='attendance')

attendance_melt.head()
```

P = present and is 100% or 1
A = absent and is 0% or 0
H = half day and is 50% or .5
T = tardy and is docked 10% so student gets 90% or .9

```python
def number_grade(grade):
    if grade == 'P':
        return 1
    elif grade == 'A':
        return 0
    elif grade == 'H':
        return .5
    elif grade == 'T':
        return .9
```

```python
# change the letter to number grade using the above function
attendance_melt['grade'] = attendance_melt['attendance'].apply(number_grade)

attendance_melt.head()
```

```python
attendance_melt.groupby('Name').grade.mean()
```

Cake Recipes

Read the cake_recipes table. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

Tidy the data as necessary.

Which recipe, on average, is the best? recipe b

Which oven temperature, on average, produces the best results? 275

Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300

```python
df_cake = pd.read_csv('cake_recipes.csv')

df_cake.head()
```

```python
df_cake = df_cake.rename(columns={'recipe:position': 'recipe'})
df_cake.head()
```

```python
df_cake[['recipe', 'rack']]= df_cake.recipe.str.split(':', expand=True)

df_cake.head()
```

```python
cake_melted = df_cake.melt(id_vars=['recipe', 'rack'], var_name= 'oven_temp', value_name='rating')

cake_melted.head()
```

```python
#which recipe is best
cake_melted.groupby('recipe').rating.mean()
```

```python
# Which oven temperature, on average, produces the best results? 275

cake_melted.groupby('oven_temp').rating.mean()
```

```python
# Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300

cake_melted.sort_values(by='rating', ascending = False).head(1)
```
